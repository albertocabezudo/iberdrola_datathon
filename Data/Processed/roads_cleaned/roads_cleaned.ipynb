```python
import geopandas as gpd

# =========================
# 1. CARGA
# =========================
df_roads = gpd.read_file(r"..\..\Raw\Roads_raw_data\260306 catalogoRCE2025.shp")

# =========================
# 2. SELECCIÓN DE COLUMNAS
# =========================
df_roads = df_roads[["provincia", "carretera", "clase", "geometry"]].copy()

# =========================
# 3. LIMPIEZA BÁSICA
# =========================

# eliminar geometrías nulas
df_roads = df_roads[df_roads["geometry"].notna()]

# eliminar duplicados
df_roads = df_roads.drop_duplicates()

# normalizar texto
df_roads["clase"] = df_roads["clase"].astype(str).str.strip().str.upper()
df_roads["carretera"] = df_roads["carretera"].astype(str).str.strip().str.upper()
df_roads["provincia"] = df_roads["provincia"].astype(str).str.strip().str.upper()

# =========================
# 4. FILTRADO DE CARRETERAS RELEVANTES
# =========================

df_roads = df_roads[
    df_roads["clase"].isin([
        "AUTOPISTA",
        "AUTOVÍA",
        "CARRETERA CONVENCIONAL"
    ])
].copy()

# =========================
# 5. FEATURE ENGINEERING
# =========================

# identificar carreteras principales
df_roads["is_highway"] = df_roads["carretera"].str.startswith(("A-", "AP-"))

# =========================
# 6. VALIDACIÓN
# =========================

print("Dimensión:", df_roads.shape)
print("\nTipos de carretera:")
print(df_roads["clase"].value_counts())

print("\nCRS:", df_roads.crs)

df_roads.head()
```
